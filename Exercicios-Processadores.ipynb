Em um processador sem a incrementação automática do CI, a lógica de controle do programa precisa incluir instruções explícitas para modificar o valor do CI quando necessário. Essas instruções podem ser chamadas de instruções de desvio (jump) ou instruções de ramificação condicional (branch) e são utilizadas para controlar o fluxo de execução do programa.

Por exemplo, suponha que temos o seguinte código em uma linguagem fictícia:

LOAD R1, A
ADD R2, R1
STORE R2, B
JUMP 5
SUB R1, R2

Nesse caso, a instrução JUMP 5 está sendo usada para alterar manualmente o valor do CI para o endereço 5. Isso indica que o próximo endereço de memória a ser buscado será o endereço 5, em vez de incrementar sequencialmente para o próximo endereço.

O programador é responsável por identificar os pontos de desvio necessários no programa e incluir instruções de desvio adequadas para modificar manualmente o valor do CI. O programa precisa ser projetado de forma apropriada, garantindo que o CI seja atualizado corretamente para direcionar o fluxo de execução para as instruções desejadas.

O Program Counter (PC) é o resgistrador que armazena o endereço da próxima instrução a ser executada, logo é muito importante para controlar a sequência de processamento das intruções. Ele se encontra dentro do processador.

(1) ADD Op: 2 ciclos de leitura da memória:
RI <- (PC): 1ª leitura na memória
Buscar o operando para somar com o valor de R0: 2ª leitura na memória

(2) SUB Op: 2 ciclos de leitura da memória:
RI <- (PC): 1ª leitura na memória
Buscar operando para subtrair com o valor do R0: 2ª leitura na memória
Armazenar o resultado no endereço do operando: 1ª Escrita na memória

(3) ADD Op1 Op2: 4 ciclos de Memória:
RI <- (PC): 1ª leitura na memória
Buscar operando 1 e 2: 2ª e 3ª leitura
Armazenar o resultado da soma do OP1 + OP2 no endereço do OP1: 1ª escrita

(4) INCR: 1 Ciclo de memória:
RI <- (PC): 1ª leitura na memória

(5) LDA Op: 2 ciclos:
RI <- (PC): 1ª leitura na memória
Buscar operando para jogar pro R0: 2ª Leitura na memória
